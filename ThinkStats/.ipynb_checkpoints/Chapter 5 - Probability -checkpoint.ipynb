# 5. Basic Probability
----
* If E represents an event, P(E) represents the probability that E will occur
* A situation where E might or might not happen is called a trial
* If we have a finite sample of n trials and we observe s successes, the probability of success is s/n.

## What about the probabilities of unique events?
* E.g. Election
   * we might like to know the probability that a candidate will win an election. 
   * But every election is unique, so there is no series of identical trials to consider.
   
* Enter <b> frequentism </b> 
    * Some people would say that the notion of probability does not apply to cases like the above. 
    * This position is sometimes called frequentism because it defines probability in terms of frequencies. If there is no set of identical trials, there is no probability.
    * However, this can be frustrating as it limits the scope of probability to physical systems that are either random or so unpredictable that we model them as random

## Going Bayesian
* Bayesianism: defines probability as a degree of belief that an event will occur
    * the notion of porbability can be applied in almost any circumstance
    * One difficulty with Bayesian probability is that it depends on a person’s state of knowledge; people with different information might have different degrees of belief about the same event.
    * So it can be more subjective than frequency probabilities

## Rules of probability
* For frequency probabilities, we can derive rules that relate probabilities of different events
    
        P(A and B) = P(A) * P(B)
        *Warning*: not always true
* `P(A and B)` is the probability that events A and B both occur
* This only applies if A and B are indepednent

### When A and B are not independent
* We can compute the conditional probability, `P(A|B)`, which is the probability of A given that we know B occurred


* From this we can derive the general relation
    
        P(A and B) = P(A) * P(B|A)
* In english:

        “The chance of both things happening is the chance that the first one happens, and then the second one given the first.”
        
* If they are independent, `P(A|B) = P(A) `   
* Because all probabilities are in the range [0,1], it is easy to show that
    
        P(A and B) <= P(A)

# 5.2 Monty Hall
* The Monty Hall problem might be the most contentious question in the history of probability
* Here is the premise:
    * Monty shows you three closed doors and tells you that there is a prize behind each door: one prize is a car, the other two are less valuable prizes like peanut butter and fake finger nails. The prizes are arranged at random.
    * The objective of the game is to guess which door has the car. If you guess right, you get to keep the car.
    * So you pick a door, which we will call Door A. We’ll call the other doors B and C.
    * Before opening the door you chose, Monty likes to increase the suspense by opening either Door B or C, whichever does not have the car. (If the car is actually behind Door A, Monty can safely open B or C, so he chooses one at random).
    * Then Monty offers you the option to stick with your original choice or switch to the one remaining unopened door.
    
* Most people have the strong intuition that there's no difference so the chance is equally 50-50 for either option
* But that is <b> wrong </b>

* Let us consider the scenarios 


* So there are 3 possible scenarios, the car being behind Door A, B, or C
* Since the prizes are arranged at random, the probability of each scenario is 1/3
> innermost circle: where the car is
> middle circle: the door you choose
> outer circle: the door monty chooses

### If you choose Door A and stick with it
* your probability of winning is Scenario A, with a probability of 1/3 
> so on the wheel it's if you choose door 1,or any scenario where you choose a Door, Stick with it and the door is behind it
    > on the wheel that is `2*3 /18 = 6/18 = 1/3`
    
### If your strategy is to switch
* If your strategy is to switch, you will win in either Scenario B or Scenario C, so the total probability of winning is 2/3.

## Exercise 5.4 Write a program that simulates the Monty Hall problem and use it to estimate the probability of winning if you stick and if you switch.

```python
import random
import numpy as np
```

```python
n = 100
STAY = 0
SWITCH = 0
for i in range(n):
    #door with the car behind it
    CAR_LOCATION = np.random.randint(3)
    
    #your initial guess
    FIRST_GUESS = np.random.randint(3)
    
    # remaining doors available
    DOORS_AVAILABLE = [x for x in [0,1,2] if x!=CAR_LOCATION and x!=FIRST_GUESS]
    # monty opens one of the other doors
    MONTY_CHOICE = DOORS_AVAILABLE[np.random.randint(len(DOORS_AVAILABLE))]
    
    # the door you switch to if you choose to switch
    SWITCH_CHOICE = [x for x in [0,1,2] if x!= MONTY_CHOICE and x!=FIRST_GUESS][0]
    
    # get the results
    if(FIRST_GUESS==CAR_LOCATION):
        STAY+=1
    if(SWITCH_CHOICE==CAR_LOCATION):
        SWITCH+=1
print('% STAY: ', STAY/n)
print('% SWITCH: ', SWITCH/n)
```

## Exercise 5.5 To understand the Monty Hall problem, it is important to realize that by deciding which door to open, Monty is giving you information.

* To see why this matters, imagine the case where Monty doesn’t know where the prizes are, so he chooses Door B or C at random.
* If he opens the door with the car, the game is over, you lose, and you don’t get to choose whether to switch or stick.
* Otherwise, are you better off switching or sticking?

```python
n = 1000
STAY = 0
SWITCH = 0
GAME_ENDS =0
for i in range(n):
    #door with the car behind it
    CAR_LOCATION = np.random.randint(3)
    
    #your initial guess
    FIRST_GUESS = np.random.randint(3)
    
    # remaining doors available
    DOORS_AVAILABLE = [x for x in [0,1,2] if x!=FIRST_GUESS]
    # monty opens one of the other doors
    MONTY_CHOICE = DOORS_AVAILABLE[np.random.randint(len(DOORS_AVAILABLE))]
    
    # the door you switch to if you choose to switch
    SWITCH_CHOICE = [x for x in [0,1,2] if x!= MONTY_CHOICE and x!=FIRST_GUESS][0]
    
    # get the results
    if(FIRST_GUESS==CAR_LOCATION):
        STAY+=1
    if(SWITCH_CHOICE==CAR_LOCATION):
        SWITCH+=1
    
    # if monty chooses the prize door, the game ends
    if(MONTY_CHOICE==CAR_LOCATION):
        GAME_ENDS+=1
        
print('% STAY: ', STAY/n)
print('% SWITCH: ', SWITCH/n)
print('% GAME ENDS: ', GAME_ENDS/n)
```

* the probability becomes more or less equal ~ 1/3 so there's no difference

# 5.3 Poincaré

* Bakery problem
    * Poincaré suspected that his local bakery was selling loaves of bread that were lighter than the advertised weight of 1 kg, so every day for a year he bought a loaf of bread, brought it home and weighed it.
    * At the end of the year, he plotted the distribution of his measurements and showed that it fit a normal distribution with mean 950 g and standard deviation 50 g. 
    * He brought this evidence to the bread police, who gave the baker a warning.
* but...
    * For the next year, Poincaré continued the practice of weighing his bread every day. At the end of the year, he found that the average weight was 1000 g, just as it should be, but again he complained to the bread police, and this time they fined the baker.
* Why?
    * the shape of the distribution was asymmetric
    * Unlike the normal distribution, it was skewed to the right, which is consistent with the hypothesis that the baker was still making 950 g loaves, but deliberately giving Poincaré the heavier ones.
    
## Exercise 5.6 Write a program that simulates a baker who 
* chooses n loaves from a distribution with mean 950 g and standard deviation 50 g, 
* and gives the heaviest one to Poincaré. 
* What value of n yields a distribution with mean 1000 g? What is the standard deviation?

```python
n_days = 365
n=5
# Poincare gets one loaf a day, so each day has its own entry over the year
POINCARE_LOAF = np.zeros(n_days)

for day in range(n_days):
    LOAF_WEIGHTS = 950 + 50 * np.random.randn(n)
    LOAF_WEIGHTS
    POINCARE_LOAF[day] = np.max(LOAF_WEIGHTS)
print("Mean: ",np.mean(POINCARE_LOAF))
print("Std: ",np.std(POINCARE_LOAF))
```

```python
MAX= 100
P_MEAN = np.zeros(MAX)
P_STD = np.zeros(MAX)
n_days = 365

for i in range(1,MAX):
    n=i 
    POINCARE_LOAF = np.zeros(n_days)
    for day in range(n_days):
        LOAF_WEIGHTS = 950 + 50 * np.random.randn(n)
        LOAF_WEIGHTS
        POINCARE_LOAF[day] = np.max(LOAF_WEIGHTS)
    P_MEAN[i]= np.mean(POINCARE_LOAF)
    P_STD[i]= np.std(POINCARE_LOAF)
```

```python
import pandas as pd
## create a dataframe
plot_df = pd.DataFrame(list(zip(list(range(len(P_MEAN))),P_MEAN,P_STD)),columns=['n','Mean','Std'])
plot_df
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# To get the value of n that yields a distribution of mean 1000 g , we can plot
fig = px.line(plot_df, x='n',y='Mean', title='Value of mean of loaf weights vs value of n', 
              labels= dict(std='Standard Deviation', mean='Mean'))
fig.update_yaxes(range=[980,1100])
fig.show()
```

# 5.4 Another rule of probability: Mutual exclusivity
* If two events are mutually exclusive, that means that only one of them can happen, so the conditional probabilities are 0:

        'P(A|B) = P(B|A) = 0'
        
* In general, the probability of A or B or both is

        P(A or B) = P(A) + P(B) - P(A-B)
* The reason we have to subtract off P(A and B) is that otherwise it gets counted twice.

## Exercise 5.8 If I roll two dice, what is the chance of rolling at least one 6?
* Scenarios
    * Both 6s
    * 1 6, other not 6
* P(both 6) = 1/6 * 1/6 = 1/36
* P(only 1 6) = P(roll 6) * P(roll not-6) = 1/6 * 5/6 = 5/36
    * Multiply by 2 as it could occur the other way round as well
    * 10/36
* P(at least 1 6) = 10/36 + 1/36 = 11/36

* Using complementary set
    * P(no 6) = 5/6 * 5/6 = 25/36
    * P(at least one 6) = 1-25/36 = 11/36

## Exercise 5.10 If you flip a coin 100 times, you expect about 50 heads, but what is the probability of getting exactly 50 heads?
* p(heads) = 1/2
* If you toss the coin 100 times, number of possible outcomes = (2^100).
* Now, for getting 50 heads in 100 tosses, number of possible outcomes = (100C50).
* So, the probability = (100C50) / (2^100).

# 5.6 Streaks and hot spots
* people usually have bad intuition regarding random processes
* if you show someone a real random sequence, they tend to see patterns where there are none
  * especially so in sports ('streaks') or having a 'hot hand'
 
 * Statisticians have tested these hypotheses in a number of sports, and the consistent result is that there is no such thing as a streak
* If you assume that each attempt is independent of previous attempts, you will see occasional long strings of successes or failures. 
* These apparent streaks are not sufficient evidence that there is any relationship between successive attempts.

## So how do we know when a cluster is meaningful?
* We can simulate the behaviour of a random system to see whether it is likely to produce a similar cluster
> this is called monte carlo simulation because generating random numbers is reminiscent of casino games (and monte carlo is known for its casino)
